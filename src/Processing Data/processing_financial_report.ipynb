```python
import pandas as pd
import glob
import os 
```

### Cân đối kế toán

```python
df = pd.read_excel(r'/home/hamornic/FinalXXX/DataScraped/Financial Report/VietstockFinance_ACB_Bao-cao-tai-chinh_BCTT_20250320-135204.xlsx', sheet_name="CDKT", skiprows=5, skipfooter=12, header=None)
column_names = ['Cân đối kế toán', 'Q1', 'Q2', 'Q3', 'Q4']
df.columns = column_names
df.head()
```

```python
folder_path = "/home/hamornic/FinalXXX/DataScraped/Financial Report V2/"

# Lấy danh sách tất cả file .xlsx trong thư mục
file_list = glob.glob(os.path.join(folder_path, "*.xlsx"))

df_list = []

for file in file_list:
    df = pd.read_excel(file, sheet_name="CDKT", skiprows=5, skipfooter=12, header=None)
    num_cols = len(df.columns)  # Số cột thực tế trong file
    column_names = ['Cân đối kế toán', 'Q1', 'Q2', 'Q3', 'Q4']  # Giữ lại các cột khác nếu có
    df.columns = column_names  
    df["stock_code"] = os.path.basename(file).split('_')[1]  # Thêm cột tên file để nhận diện
    df_list.append(df)
    # print(file) if num_cols != 5 else num_cols
final_df = pd.concat(df_list, ignore_index=True)
```

```python
final_df.head(30)
```

```python
# Tạo cột "Năm", ban đầu là NaN
final_df["Năm"] = None

# Xác định các dòng chứa thông tin năm (các dòng có chứa "Cân đối kế toán")
for idx, row in final_df.iterrows():
    if "Cân đối kế toán" in str(row["Cân đối kế toán"]):
        # Trích xuất năm từ cột "Q1" (giả sử đúng định dạng "Q1/2017")
        year = row["Q1"].split("/")[1] if isinstance(row["Q1"], str) and "/" in row["Q1"] else None
    if year:
        final_df.at[idx, "Năm"] = year  # Gán năm cho dòng chứa "Cân đối kế toán"
```

```python
final_df.tail(50)
```

```python
keywords = ["Giai đoạn", "Hợp nhất", "Kiểm toán", "Công ty kiểm toán", "Ý kiến kiểm toán"]
final_df = final_df[~final_df["Cân đối kế toán"].astype(str).isin(keywords)]
```

```python
final_df = final_df.dropna(subset=["Cân đối kế toán"]) 
```

```python
final_df = final_df.sort_values(by=["stock_code", "Năm"])
```

```python
final_df.to_csv(r"/home/hamornic/FinalXXX/Financial Report Processed/cdkt.csv", index=False, encoding="utf-8-sig")
```

### Chỉ số tài chính

```python
folder_path = "/home/hamornic/FinalXXX/DataScraped/Financial Report V2/"

# Lấy danh sách tất cả file .xlsx trong thư mục
file_list = glob.glob(os.path.join(folder_path, "*.xlsx"))

df_list = []

for file in file_list:
    df = pd.read_excel(file, sheet_name="CSTC", skiprows=5, skipfooter=12, header=None)
    num_cols = len(df.columns)  # Số cột thực tế trong file
    column_names = ['Chỉ số tài chính', '','period_1', 'period_2', 'period_3', 'period_4']  # Giữ lại các cột khác nếu có
    df.columns = column_names  
    df["stock_code"] = os.path.basename(file).split('_')[1]  # Thêm cột tên file để nhận diện
    df_list.append(df)
    # print(file) if num_cols != 5 else num_cols
final_df = pd.concat(df_list, ignore_index=True)
```

```python
final_df.columns
```

```python
period_1 = final_df[['Chỉ số tài chính', 'period_1', '', 'stock_code']]
period_2 = final_df[['Chỉ số tài chính', 'period_2', '', 'stock_code']]
period_3 = final_df[['Chỉ số tài chính', 'period_3', '', 'stock_code']]
period_4 = final_df[['Chỉ số tài chính', 'period_4', '', 'stock_code']]
```

```python
period_2.rename(columns={'period_2':'period_1'}, inplace=True)
period_3.rename(columns={'period_3':'period_1'}, inplace=True)
period_4.rename(columns={'period_4':'period_1'}, inplace=True)
```

```python
final_df = pd.concat([period_1, period_2, period_3, period_4], ignore_index=True)
final_df.head()
```

```python
# Tạo cột "Năm", ban đầu là NaN
final_df["Năm"] = None

# Xác định các dòng chứa thông tin năm (các dòng có chứa "Cân đối kế toán")
for idx, row in final_df.iterrows():
    if "Chỉ số tài chính" in str(row["Chỉ số tài chính"]):
        # Trích xuất năm từ cột "Q1" (giả sử đúng định dạng "Q1/2017")
        year = row["period_1"]
    if year:
        final_df.at[idx, "Năm"] = year 
```

```python
keywords = ["Giai đoạn", "Hợp nhất", "Kiểm toán", "Công ty kiểm toán", "Ý kiến kiểm toán"]
final_df = final_df[~final_df["Chỉ số tài chính"].astype(str).isin(keywords)]
final_df = final_df.dropna(subset=["Chỉ số tài chính"]) 
final_df = final_df.sort_values(by=["stock_code", "Năm"])
```
