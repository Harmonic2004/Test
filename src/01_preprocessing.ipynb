#### Missing and Outlier

```python
from statsmodels.tsa.seasonal import MSTL
from statsmodels.tsa.stattools import adfuller
```

```python
data = stock_dfs['ACB']
data[data['transaction_date'].dt.year == 2018].head()
```

```python
def process_history(df):
    all_dates = pd.date_range(start=df['transaction_date'].min(), end=df['transaction_date'].max())
    existing_dates = set(df['transaction_date'])


    missing_holidays = [d for d in all_dates if d not in existing_dates and d.weekday() < 5]

    holiday_df = pd.DataFrame({'transaction_date': missing_holidays})
    holiday_df['is_history'] = 1  


    df = pd.concat([df, holiday_df], ignore_index=True).sort_values(by='transaction_date')

    df.fillna(method='ffill', inplace=True)
    df['is_history'] = df['is_history'].fillna(0).astype(int)
    df.reset_index(drop=True, inplace=True)

    return df
```

```python
def split_data(data, start_date = "2018-01-01", end_date = "2024-12-31"):
    mask = (data['transaction_date'] >= start_date) & (data['transaction_date'] <= end_date)
    filtered_data = data.loc[mask]
    return filtered_data
```

```python
def process_missing_data(data):
    print(f"Have {data['closing_price'].isna().sum()} miss data")
    data['closing_price'] = data['closing_price'].interpolate(method="linear")
    data['closing_price'] = data['closing_price'].fillna(method='ffill').fillna(method='bfill')

    mstl = MSTL(data['closing_price'], periods=[5])
    res = mstl.fit()
    
    df_deseasonal = data["closing_price"] - res.seasonal
    df_deseasonal_imputed = df_deseasonal.interpolate(method="linear")
    df_deseasonal_imputed = df_deseasonal_imputed.fillna(method='ffill').fillna(method='bfill')

    data['closing_price'] = df_deseasonal_imputed + res.seasonal
    data['closing_price'] = data['closing_price'].round(1)
    return data
    return data
```

```python
def process_outlier_data(data, visualize=False):
    mstl = MSTL(data['closing_price'].interpolate(method="linear"), periods=(5))
    res = mstl.fit()
    data['resid'] = res.resid
    Q1 = data['resid'].quantile(0.25)
    Q3 = data['resid'].quantile(0.75)
    IQR = Q3 - Q1

    alpha = 3
    data['lower'] = Q1 - alpha * IQR
    data['upper'] = Q3 + alpha * IQR
    data['is_outlier'] = (data['resid'] < data['lower']) | (data['resid'] > data['upper'])
    print(f"Have {data['is_outlier'].sum()} outlier data")
    data['original_price'] = data['closing_price'].copy()
    data.loc[data['is_outlier'], 'closing_price'] = np.nan
    data = process_missing_data(data)
    # print(data["closing_price"])
    if visualize:
        # Tạo đối tượng biểu đồ
        fig = go.Figure()

        # Thêm dữ liệu và đường biểu đồ
        fig.add_trace(go.Scatter(x=data["transaction_date"], y=data["closing_price"], mode='lines', name='Giá'))
        fig.add_trace(go.Scatter(x=data["transaction_date"], y=data["original_price"], mode='lines', line=dict(color="orange"), name='Giá ban đầu'))
        fig.add_trace(go.Scatter(x=data["transaction_date"], y=res.resid, mode='lines', name='Phần dư'))
        fig.add_trace(go.Scatter(x=data["transaction_date"][data['is_outlier']], y=data['closing_price'][data['is_outlier']], mode='markers', marker=dict(color='red'), name='Ngoại lệ'))
        fig.add_trace(go.Scatter(x=data["transaction_date"][data['is_outlier']], y=data['resid'][data['is_outlier']], mode='markers', marker=dict(color='red'), name='Ngoại lệ Resid'))
        fig.add_trace(go.Scatter(x=data["transaction_date"], y=data['upper'], mode='lines', line=dict(color='black', dash='dash'), name='Ngưỡng trên'))
        fig.add_trace(go.Scatter(x=data["transaction_date"], y=data['lower'], mode='lines', line=dict(color='black', dash='dash'), name='Ngưỡng dưới'))

        # Đặt nhãn cho trục x và trục y
        fig.update_xaxes(title_text='Thời gian')
        fig.update_yaxes(title_text='Giá')

        # Đặt tiêu đề cho biểu đồ
        fig.update_layout(title='Biểu đồ giá và các giá trị Outliers')

        # Hiển thị biểu đồ
        fig.show()

    return data
```

```python
def save_file(data, file_path):
    data.drop(columns=["resid", "lower", "upper", "is_outlier", "original_price"], inplace=True)
    data.to_csv(rf'{file_path}', index=False, encoding='utf-8-sig')
```

```python
df_final = pd.DataFrame()
for key, df in stock_dfs.items():
    print(f"Processing {key}")
    df = split_data(df, start_date="2017-01-01", end_date="2024-12-31")
    if df.shape[0] < 400:
        continue
    print(df.isna().sum().values.sum()  )
    df = process_history(df)
    df = process_missing_data(df)

    df = process_outlier_data(df)
    df["stock_code"] = key
    df_final = pd.concat([df_final, df])

    
    
```
